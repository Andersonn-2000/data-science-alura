```python
""""Utilizando o conjunto com os dados de localização das 
estações de metrô obtido no portal DataRio (utilizadas no último vídeo),
identifique as quatro estações de metrô mais próximas e as quatro estações
de metrô mais distantes dos quatro pontos abaixo:"""

url_m = 'https://github.com/allanspadini/curso_geopandas/raw/main/dados/metro.geojson'
subway = gpd.read_file(url_m)
```

```python
# Criando um DataFrame com os dados dos problemas
points = {
    'Latitude':[-22.9531, -23.0278, -22.8463, -22.8969],
    'Longitude':[-43.1884, -43.4665, -43.3007, -43.3165]
}

data_base = pd.DataFrame(points)
```

```python
# Convertendo o DataFrame em um GeoDataFrame
geometry = [Point(x) for x in zip(data_base.Latitude, data_base.Longitude)]
crs = {'init':'epsg:4326'}
geo_dados = gpd.GeoDataFrame(data_base, crs=crs, geometry=geometry)
```

```python
# Modificando o CRS dos GeoDataFrames
geo_dados = geo_dados.to_crs('+proj=utm +zone=23 +south +ellps=GRS80 +towgs84=0,0,0,0,0,0,0 +units=km +no_defs')
subway = subway.to_crs('+proj=utm +zone=23 +south +ellps=GRS80 +towgs84=0,0,0,0,0,0,0 +units=km +no_defs')
```

```python
# Criando colunas com as respostas
geo_dados['Metro_Perto'] = geo_dados['geometry'].apply(
    lambda x: subway['nome'].iloc[subway.distance(x).idxmin()]
)

geo_dados['Dist_min'] = geo_dados['geometry'].apply(
    lambda x: "{0:.2f} km".format(subway.distance(x).min())
)

geo_dados['Metro_Longe'] = geo_dados['geometry'].apply(
    lambda x: subway['nome'].iloc[subway.distance(x).idxmax()]
)

geo_dados['Dist_max'] = geo_dados['geometry'].apply(
    lambda x: "{0:.2f} km".format(subway.distance(x).max())
)

geo_dados[['Metro_Perto', 'Dist_min', 'Metro_Longe', 'Dist_max']]
```

```python
# Adicionando o CRS ás bases de dados
train = train.to_crs('+proj=utm +zone=23 +south +ellps=GRS80 +towgs84=0,0,0,0,0,0,0 +units=km +no_defs')
brt = brt.to_crs('+proj=utm +zone=23 +south +ellps=GRS80 +towgs84=0,0,0,0,0,0,0 +units=km +no_defs')
```

```python
#Verificando se os dados estão corretamente dentro do mapa
train = train[train.within(rj_municipality.iloc[0].geometry)]
brt = brt[brt.within(rj_municipality.iloc[0].geometry)]
```

```python
# Criar mapa de disposição dos dados
base = rj_municipality.plot(color='white', edgecolor='black')
geo_data.plot(ax=base, alpha=0.2)
```

```python
# Criaando mapa de calor da correlação dos dados
transport = pd.concat([train.geometry, subway.geometry, brt.geometry], ignore_index=True)
```

```python
# Distância minima até o meio de transporte
geo_data['Dist_transporte'] = geo_data['geometry'].apply(
    lambda x: transport.distance(x).min()
)
```

```python
# Calculando a correlação
geo_data_num = geo_data.select_dtypes(include=[np.number])
correlation = geo_data_num.corr()
```

```python
sns.heatmap(correlation, annot=True, cmap='RdYlBu', linewidths=0.5)
```

```python
""""A distância dos imóveis até as praias influenciam no preço dos mesmos?"""

url_p = '../data/Cobertura_Vegetal_e_Uso_da_Terra_2018.geojson'
```

```python
beachs = gpd.read_file(url_p)
beachs.head()
```

```python
beachs = beachs.copy()
```

```python
beachs = beachs.to_crs('+proj=utm +zone=23 +south +ellps=GRS80 +towgs84=0,0,0,0,0,0,0 +units=km +no_defs')
beachs.plot(color='white', edgecolor='black')
```

```python
beachs['classe'].unique()
```

```python
# Localização das praias no municipio do RJ
beachs = beachs[beachs['classe'] == 'Praia']
base = rj_municipality.plot(color='white', edgecolor='black', alpha=0.4)
beachs.plot(ax=base, color='white', edgecolor='blue')
```

```python
# Calculando a menor distância até as praias
geo_data['Dist_Praia'] = geo_data['geometry'].apply(
    lambda x: beachs.distance(x).min()
)
```

```python
""""Calculando a correlação entre a distância das praias e os preços dos imóveis"""
corr = geo_data.select_dtypes(include=[np.number])
correlation = corr.corr()
```

```python
# Plotando o heatmap
sns.heatmap(correlation, annot = True, cmap='RdYlBu', linewidths=0.5)
```

#### Obtemos algo semelhante ao que observamos nos outros meios de transporte, sendo uma correlação de -0.36. Assim, quanto mais próximo da praia, maior é o valor do imóvel.

```python
# Plotando a distância da praia em relação ao valor
sns.scatterplot(x='Dist_Praia', y='Valor', data=geo_data)
```

#### Como resultado, temos um gráfico do valor em relação à distância da praia. Dessa forma, analisamos que quando a distância da praia é muito próxima de zero ou até aproximadamente 2.5km, temos o valor dos imóveis dramaticamente aumentando.
